```python
# creating the function:
def get_region(row):
    states = {
        'AK': 'Other','AL': 'South','AR': 'South','AS': 'Other','AZ': 'West','CA': 'West','CO': 'West','CT': 'East','DC': 'East',
        'DE': 'East','FL': 'South','GA': 'South','GU': 'Other','HI': 'Other','IA': 'Midwest','ID': 'West','IL': 'Midwest','IN': 'Midwest','KS': 'Midwest',
        'KY': 'South','LA': 'South','MA': 'East','MD': 'East',
        'ME': 'East','MI': 'Midwest',
        'MN': 'Midwest','MO': 'Midwest','MP': 'Other',
        'MS': 'South','MT': 'West','NA': 'Other','NC': 'South','ND': 'Midwest','NB': 'Midwest','NH': 'East','NJ': 'East','NM': 'West','NV': 'West','NY': 'East',
        'OH': 'Midwest','OK': 'South','OR': 'West','PA': 'East','PR': 'Other','RI': 'East',
        'SC': 'South','SD': 'Midwest','TN': 'South','TX': 'South','UT': 'West','VA': 'South','VI': 'Other',
        'VT': 'East','WA': 'West','WI': 'Midwest','WV': 'South','WY': 'West'}
    list=[]
    for i in row:
        if i[0:2] in states.keys():
            list.append(states.get(i[0:2]))  
        else:
            list.append("lost")
    return list
```
